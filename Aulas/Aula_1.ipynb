### Desafio 03: Adicionar título no gráfico

```
import matplotlib.pyplot as plt
```

```
dados['NU_IDADE'].hist(bins=20, figsize = (10,8))
plt.title('Histograma da idade dos inscritos no ENEM 2019', fontdict = {'fontsize' : 20})
plt.show()
```

### Desafio 04: Plotar os histogramas das idades dos treineiros e não treineiros

```
plt.figure(figsize = (10,10))
plt.subplot(2,1,1)
dados[dados['IN_TREINEIRO'] == 1]['NU_IDADE'].hist(bins=20)
plt.title('Idades dos TREINEIROS', fontdict = {'fontsize' : 20})
plt.subplot(2,1,2)
dados[dados['IN_TREINEIRO'] == 0]['NU_IDADE'].hist(bins=20)
plt.title('Idades dos NÃO TREINEIROS', fontdict = {'fontsize' : 20})
plt.show()
```

### Desafio 05: Comparar as distribuições das provas em inglês e espanhol

```
plt.figure(figsize = (16,10))

plt.subplot(2,2,1)
dados[dados['TP_LINGUA'] == 1]['NU_NOTA_LC'].plot.box(grid=True)
plt.title('Boxplot Espanhol', fontdict = {'fontsize':16})

plt.subplot(2,2,2)
dados[dados['TP_LINGUA'] == 0]['NU_NOTA_LC'].plot.box(grid=True)
plt.title('Boxplot Inglês', fontdict = {'fontsize':16})

plt.subplot(2,2,3)
dados[dados['TP_LINGUA'] == 1]['NU_NOTA_LC'].hist(bins = 20)
plt.title('Histograma Espanhol', fontdict = {'fontsize':16})

plt.subplot(2,2,4)
dados[dados['TP_LINGUA'] == 0]['NU_NOTA_LC'].hist(bins = 20)
plt.title('Histograma Inglês', fontdict = {'fontsize':16})

plt.show()
```
